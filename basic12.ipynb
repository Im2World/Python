```python
#파이썬 기초12강 - 파일다루기
 #try - catch - finally 문 실행
    
f=0 # 변수 선언 및 값 0 초기화
try: # 정상 실행구문
    f=open("LICENSE.txt","rt") # 해당이름의 텍스트파일을 읽기모드로 열기
    while True: # 무한 루프
        r=f.readline() # 텍스트 파일에서 한 줄 읽고 r에 담기
        if not r: # 라인 없으면
            break # 반복 종료
        print(r) # 한줄씩 파일 내용 출력
        
except FileNotFoundError: # 예외 에러 처리
    print("파일없음") # 파일이 없으면, 파일없음 메시지 출력
    
finally: # 무조건 실행 => 마지막 실행구문
    if(f): f.close() # 파일 닫고
    print("프로그램 끝") # 프로그램끝 메시지 출력
```

```python
#파일 쓰기
# 100명의 국어,영어,수학 점수를 임의로 만든 후 csv형태의 파일의 기록하는 예제
import random as r # 랜덤 모듈 불러오고 r이라고 명명

f=0 # 변수 선언 및 값 0 초기화
f=open("score.txt","wt") # 텍스트 파일을 쓰기로 열기 => 파일이름은 score.txt로 지정

# 아래와 같은 형태를 만들기 위하여 str.format 함수를 사용할 수 있다. 
#   one_person="{0:s},{1:d},{2:d},{3:d}\n"
#.format("홍길동"+str(i),r.randrange(0,100),r.randrange(0,100),r.randrange(0,100)) 

#txt파일에 값 채워넣는 반복문
for i in range(100): # 변수 i 범위는 0~99까지로 총 100회 반복
    
    # 홍길동+번호,국어점수,영어점수,수학점수 랜덤 생성해 one_person에 한 줄로 대입
    #각 과목 점수 범위는 0~99
    one_person="홍길동"+str(i)+","+str(r.randrange(0,100))+","+str(r.randrange(0,100))+"," \
    +str(r.randrange(0,100))+"\n" 
    f.write(one_person) #한 줄씩 쓴다
    print(one_person) # 한명씩 [홍길동+번호,국어점수,영어점수,수학점수] 출력
f.close() # 파일 닫기
```

```python
#파일 읽기
#앞서 만든 csv파일을 한 줄씩 읽어서 처리하는 예제
#csv파일 자료구조 : 홍길동99,10,35,78  => 이름+번호, 국어성적,영어성적,수학성적 순

f=0# 변수 선언 및 값 0 초기화
f=open("score.txt","rt")  # score.txt라는 텍스트 파일을 읽기모드로 열기

while True: # 무한 루프
    one_person=f.readline() # 텍스트 파일을 한 줄씩 읽어서 one_person에 대입
    if not one_person: break # 읽을 내용이 없으면 루프 종료
        
    one_person=one_person.replace("\n","") # 한 줄씩 읽은 one_person에 줄바꿈이 있으면 제거함. 
    item=one_person.split(",") #one_person을 콤마를 구분자로 잘라서 변수 item에 대입
    print("="*15) # 구분선으로 등호 15개 출력
    print("이름: ", item[0]) # 이름 출력 => item인덱스 0번 값
    print("국어: ", item[1]) # 국어 성적 출력 => item인덱스 1번 값
    print("영어: ", item[2]) # 영어 성적 출력 => item인덱스 2번 값
    print("수학: ", item[3]) # 수학 성적 출력 => item인덱스 3번 값
    print("="*15) # 구분선으로 등호 15개 출력
f.close() # 파일 닫기
```

```python
#파이썬에서 sqlite3 사용 실습 
#테이블 생성, 삭제
#파이썬에서 DB다루는 경우 처리순서
#connect - cursor - execute -  {fetchall} - cursor.close - con.close

import sqlite3 as s # sqlite3 사용위해 임포트 후 s라는 명칭으로 사용

con=s.connect("my01.db") # DB연결, 현재 소스파일의 디렉토리
cursor = con.cursor() #데이터베이스 사용공간 생성

#table 삭제 
#cursor.execute("drop table examtable;")   #이 소스코드로 해당db처음 생성이라 제거할 테이블없음
con.commit() #버퍼에 처리된 내용을 기록commit

# table 생성 : 테이블명 examtable, 항목은 name varchar(20),studentid int not null primary key kor int,eng int,mat int
cursor.execute("create table examtable(name varchar(20),studentid int not null primary key,\
kor int,eng int,mat int)")
con.commit()  #버퍼에 처리된 내용을 기록commit
cursor.close()
con.close() # 닫기
```

```python
#데이터 선택(select)
#select는 fetch를 사용하여 데이터를 반복으로 가져옴

import sqlite3 as s  # sqlite3 사용위해 임포트 후 s라는 명칭으로 사용
con=s.connect("my01.db")  # 파일 연결
cursor = con.cursor() #데이터베이스 사용공간 생성

cursor.execute("select * from examtable") #examtable 테이블의 모든 데이터 가져옴

one_rec=cursor.fetchall()  #해당 데이터를 전부 가져옴

for field in one_rec:  #해당 데이터를 반복문을 통해 하나씩 처리 
    kor=field[2]  #필드 값을 변수에 대입
    eng=field[3]  #필드 값을 변수에 대입 
    mat=field[4]  #필드 값을 변수에 대입
    
    #출력
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d"\
          %(field[0],field[1],kor,eng,mat,kor+eng+mat,int((kor+eng+mat)/3))) 

cursor.close()
con.close() # 닫기
```
