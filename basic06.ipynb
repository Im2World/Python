```python
#파이썬 기초06강 - 반복문
#while반복문
i=1 #i에 1 대입
while i < 5:    #i가 5보다 작으면 계속 반복
    i+=1    #i 값 1증가
    print(i) #출력
```

```python
#while반복문 - 누적구하는 예제1
#카운터 증가 후 누적 => 1~100까지 합 계산
i=0 #변수 i에 0 대입
sum=0   #변수 sum에 0 대입 => 합계 누적할 변수
while i <100: #i가 100보다 작으면 계속 반복 => 0~99까지 실행횟수 100회
    i+=1 #i 값 1 증가 (0+1 , 1+1, 2+1, 3+1 ...) => i는 계속 1씩 증가 => 12345678...
    sum = sum + i #sum은 sum에 i 더한 값 누적 => (1,2,3,4,5,6,... 99,100까지 더함)
print("sum=",sum) #while 반복문 종료 후 출력
```

```python
#제어문
#continue : 해당 문장 아래를 실행하지않고 다시 루프를 실행할 경우 
for i in range(10):  #i는 0부터 9까지
    if(i==3): #i가 3이면
        continue    #해당 문장의 아래 코드는 실행하지 않고 반복문의 맨 처음으로 이동해 반복수행
    print(i,",",end="")  #한 줄에 출력
```

```python
#제어문 - 별찍기
#i가 30의 배수일 때 줄바꿈을 하므로 모든 줄의 별 개수를 통일하기 위해 range stop값은 121 => 1~120까지 반복
for i in range(1,121): #변수 i는 1~(121-1)까지 반복
    print("*",end="") #문자 "*" 출력 후 end로 마지막문자 지정 => 한 줄에 * 계속 출력
    if(i%30==0): print(end="\n") #i가 30의 배수면, end마지막 문자는 줄바꿈 => 30배수일 때 줄바꿈
```

```python
#다중반복구조 - 구구단 예제
for i in range(1,10):  #반복문1 : 변수 i는 1~(10-1)까지 반복
    for j in range(1,10): #반복문2 : 변수 j는 1~(10-1)까지 반복 
        print( i,"*",j,"=",i*j) #출력 : 문자 "i * j = "과 i와 j 곱한 결과
    print("----------") #구분선
```

```python
#다중반복구조 - 역직각삼각형1
for i in range(1,10):  #반복문1 : 변수 i는 1~(10-1)까지 반복
    for j in range(i,10):  #반복문2 : 변수 j는 i~(10-1)까지 반복 => 첫줄 별 9개, 두번째줄 별 8개 ...
        print( "*",end="") #반복문 2안에서 문자 "*" 출력 후 end로 마지막문자 지정 => 한 줄에 * 계속 출력
    print(end="\n") #반복문2 수행 후 줄바꿈
    #print("")      #이렇게 입력해도 위와 같은 결과
```

```python
#다중반복구조 - 역직각삼각형2
#다중반복구조 - 역직각삼각형1을 무한루프로 수정
i=1 #변수 i에 1대입 => 반복문 시작 값 지정
while True: #반복문1 : 무한루프
    j=i #변수 j에 값 i대입, j+=1로 1씩 증가했던 j에 위에서 받은 i값을 대입함.
    while True: #반복문2: 무한루프
        if(j==10): #j가 10이면
            break  #반복문2 무한루프 종료
        print( "*",end="") #j가 10이 아니면 한 줄에 * 출력,
        j+=1  #j값 1증가 => 1,2,3,4,5,6,7,8,9 만큼 * 출력
    if(i==10): #i가 10이면,
        break #반복문1 무한루프 탈출=> 반복 종료
    print(end="\n") #줄바꿈
    i+=1 #i 값 1증가
```

```python
#다중반복구조 - 심화실습1 - 달력출력
#윤달없는 것으로 가정
Mon=1 # 변수 Mon의 초기값은 1 => 마지막 소스코드에서 Mon를 1씩 증가 시킴 => LsatDay 리스트 마지막 인덱스까지 반복 => 1~12
#반복문1
for LastDay in [31,28,31,30,31,30,31,31,30,31,30,31]: #변수 LastDay값은 매 월의 마지막 일자 담은 배열에서 순서대로 대입
    print(Mon,"월 :",end="") #현재 월 출력
    #반복문2
    for Day in range(1,LastDay+1): #변수 Day는 1~LastDay까지 => range는 끝값 -1범위까지 반복
        print(Day,end=",")  #매월 1일부터 마지막일까지 콤마로 구분해서 한 줄에 출력
    print(end="\n") #반복문2 종료 후 줄바꿈 => 다음달을 다음 줄에 출력
    Mon+=1 #Mon 값 1증가 => 반복문 12회 돌고, Mon의 최종값은 13
```

```python
#다중반복구조 - 심화실습2 - 구구단 출력
#한 줄에 2,3,4단 출력
#한 줄에 3개단 출력
for i in range(2,10,3): #반복문1 : 변수 i는 2~9까지 3씩 증가 => 세로 2단, 5단, 8단 줄바꿔 출력
    for j in range(2,10): #반복문2 : 변수 j는 2~9까지 1씩 증가 => 구구단에서 뒷자리에 곱하는 값 
        print(i,"*",j,"=",i*j,end="\t") #문자 "i*j=" 와 실제 i*j결과 출력, 마지막문자로 탭 넣어서 한 줄에 출력
        print(i+1,"*",j,"=",i*j,end="\t") #위와 같은 줄에 i+1 단 출력, 마지막문자로 탭 넣어서 한 줄에 출력
        if(i+2!=10): #i+2가 10이 아니면,
            print(i+2,"*",j,"=",i*j,end="\n") #위와 같은 줄에 i+2단 출력 후 줄바꿈!
        else:   #i+2가 10이면
            print(end="\n") #10단은 출력안하고 줄바꿈
    print(end="\n") #줄바꿈
```

```python
#다중반복구조 - 심화실습3 - 구구단 출력2
#한 줄에 2,5,8단 출력=> 3,6,9단 출력 => 4,7단 출력
for i in range(2,5): #반복문1 : 변수 i는 2~4까지 1씩 증가 => 세로줄 2단, 3단, 5단
    for j in range(2,10): #반복문2 : 변수 j는 2~ 9까지 1씩 증가 => 가로줄 구구단 뒷자리(곱해주는 수) 변화
        print(i,"*",j,"=",i*j,end="\t") #문자 "i*j=" 와 실제 i*j결과 출력, 마지막문자로 탭 넣어서 한 줄에 출력
        print(i+3,"*",j,"=",i*j,end="\t") #한 줄에 2,5,8단 출력 => 3,6,9단 출력 => 4,7단 출력
        if(i+6!=10): print(i+6,"*",j,"=",i*j,end="\n") #각 줄 우측 끝단(3번째단)이 10이아니라면 출력
        else: print(end="\n") #각 줄 우측 끝단(3번째단)이 10이면 출력하지 않고 줄바꿈
    print(end="\n") #줄바꿈
```
